```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Medium','High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df.columns
```

```python
# Examine the unique categories in categories column
categorical_cols = df[['Nationality','Fee Structure','Loyalty Classification',
                       'Amount of Credit Cards','Risk Weighting','BRId','GenderId','IAId','Properties Owned','Income Band']]

for col in categorical_cols.columns:
    print(f"The value counts for {col} is")
    display(df[col].value_counts())
```

## univariate analysis

```python
for i,predictor in enumerate(df[['Nationality','Fee Structure','Loyalty Classification','Amount of Credit Cards','Risk Weighting','BRId','GenderId','IAId','Properties Owned','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

## Bivariate analysis

```python
for i,predictor in enumerate(df[['Nationality','Fee Structure','Loyalty Classification','Amount of Credit Cards','Risk Weighting','BRId','GenderId','IAId','Properties Owned','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 
                  'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# univariate analysis
plt.figure(figsize=(20,20))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
df['Estimated Income'].plot(kind='box')
```

## heatmap

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,5))

sns.heatmap(correlation_matrix,cmap='crest',fmt='.2f',annot=True)
plt.title("Heat map for numerical columns")
plt.show()
```

## EDA insights
1. There is a strong correlation between "Bank deposits" with "Checking Accounts","Savings account" and "Foreign Currency account"
   which means customers have high funds on one account will have funds in other accounts.
2. Most of the customers are from Europe, Asia and America
